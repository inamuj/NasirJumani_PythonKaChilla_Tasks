```python
# import libraries
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dropping few columns and make a new data set

new_hararat = hararat.drop(['Domain Code', 'Element Code', 'Year Code', 'Months Code','Flag','Flag Description' ], axis=1)
new_hararat.head()
```

```python
new_hararat.mean()
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x="Year", y="Value", hue="Area",data=new_hararat).set(title="Consolidated overview of Climate change across SAARC region during 1961-2021")
```
